```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
df.head()
```

```python
dictionary = {'setosa':0, 'versicolor':1, 'virginica':2}
df['species'] = df['species'].map(dictionary)
```

```python
df.head()
```

```python
x = df.drop('species',axis=1)
```

```python
import tensorflow as tf
```

```python
y = tf.keras.utils.to_categorical(df['species'])
```

```python
inputlayer = tf.keras.layers.Dense(units=4, input_dim = 4)
hlayer1 = tf.keras.layers.Dense(units=32)
hlayer2 = tf.keras.layers.Dense(units=32)
hlayer3 = tf.keras.layers.Dense(units=64)
hlayer4 = tf.keras.layers.Dense(units=64)
hlayer5 = tf.keras.layers.Dense(units=128)
hlayer6 = tf.keras.layers.Dense(units=64)
outputlayer = tf.keras.layers.Dense(units=3, activation='softmax')

```

```python
model = tf.keras.Sequential()
model.add(inputlayer)
model.add(hlayer1)
model.add(hlayer2)
model.add(hlayer3)
model.add(hlayer4)
model.add(hlayer5)
model.add(hlayer6)
model.add(outputlayer)

```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=20)
```

```python
df.head()
```

```python
import numpy as np
```

```python

input_data = np.array([[5.1,3.5,1.4,0.2]])
prediction = model.predict(input_data)
print(prediction)
```
